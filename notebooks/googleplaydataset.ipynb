```python

df_copy['Size'] = df_copy['Size'].str.replace('M', '000', regex=False)
df_copy['Size'] = df_copy['Size'].str.replace('k', '', regex=False)
df_copy['Size'] = df_copy['Size'].replace(['Varies with device', 'Variable with device'], np.nan)
df_copy['Size'] = pd.to_numeric(df_copy['Size'], errors='coerce')

```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Price'].unique()
```

```python
chars_to_remove=['$',',','+']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df_copy[cols]=df_copy[cols].str.replace(item,'')
```

```python
df_copy['Price'].unique()
```

```python
df_copy['Installs'].unique()
```

```python
df_copy['Installs']=df_copy['Installs'].astype('int')
df_copy['Price']=df_copy['Price'].astype('float')
```

```python
df_copy['Last Updated']=pd.to_datetime(df_copy['Last Updated'])
df_copy['Day']=df_copy['Last Updated'].dt.day
df_copy['Month']=df_copy['Last Updated'].dt.month
df_copy['Year']=df_copy['Last Updated'].dt.year
```

```python
df_copy.info()
```

```python
print(df_copy['Type'].unique())
print(df_copy['Content Rating'].unique())
```

```python
# Remove rows where 'Type' isn't Free or Paid
df_copy = df_copy[df_copy['Type'].isin(['Free', 'Paid'])]
```

```python
df_copy
```

```python

df_copy = df_copy.drop_duplicates()
```

```python
df_copy = df_copy.sort_values(by='Reviews', ascending=False)
df_copy = df_copy.drop_duplicates(subset=['App'], keep='first')
```

```python
df_copy['Size'] = df_copy['Size'].fillna(df_copy['Size'].median())
```
